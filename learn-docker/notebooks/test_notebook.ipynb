### Approximating value of PI using Monte-Carlo

```python
import random, math, PIL, IPython.display
from PIL import ImageDraw
```

```python
def approximate_pi(trials):
    ymax = 300
    xmax = 300
    
    inside = 0
    img = PIL.Image.new('RGB', (xmax, ymax), "black")
    draw = ImageDraw.Draw(img)
    for i in range(trials):
        x = random.random()
        y = random.random()
        
        if math.hypot(x, y) < 1:
            inside+=1
            draw.point((int(x*xmax), int(y*ymax)), fill="green")
        else:
            draw.point((int(x*xmax), int(y*ymax)), fill="red")
    img.save('monte-carlo-pi.png')
    
    pi = 4.0 * inside/trials
    return ('trials = {}  π = {}').format(trials, pi)
```

```python
approximate_pi(100000)
```
